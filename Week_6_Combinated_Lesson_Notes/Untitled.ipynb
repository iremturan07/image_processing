```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_img = cv2.imread("./aslan.jpg") 
```

```python
my_img_2 = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(my_img_2)
```

```python
my_img_vertical = cv2.flip(my_img_2, 0)
my_img_horizontal = cv2.flip(my_img_2, 1)
my_img_both = cv2.flip(my_img_2, -1)

```

```python
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2)
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_img_vertical)
plt.axis('off')
plt.title("VERTICAL Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_img_horizontal)
plt.axis('off')
plt.title("HORIZONTAL Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_img_both)
plt.axis('off')
plt.title("BOTH Image")
```

```python
print(f"görüntünün boyut bilgileri = {my_img_2.shape}")
print("-"*50)
print(f"görüntünün yüksekliği = {my_img_2.shape[0]}")
print(f"görüntünün genişliği = {my_img_2.shape[1]}")
```

```python
my_resized_img = cv2.resize(my_img_2, (600,450), cv2.INTER_LANCZOS4)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_2)
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_resized_img)
plt.axis('off')
plt.title("Resized Image")
```
