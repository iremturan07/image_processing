```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
my_original_img = cv2.imread("./aslan.jpg",0)
```

```python
_, my_thresh_binary_img = cv2.threshold(my_original_img, thresh=128, maxval=255, type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")
```

```python
_, my_thresh_binary_inv_img = cv2.threshold(my_original_img, thresh=128, maxval=255, type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(14, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING INV Image")
```

```python
_, my_thresh_tozero_img = cv2.threshold(my_original_img, thresh=128, maxval=255, type=cv2.THRESH_TOZERO)
```

```python
fig = plt.figure(figsize=(14, 7))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING INV Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_tozero_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING TOZERO Image")
```

```python
_, my_thresh_tozero_inv_img = cv2.threshold(my_original_img, thresh=128, maxval=255, type=cv2.THRESH_TOZERO_INV)
```

```python
fig = plt.figure(figsize=(14, 7))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING INV Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_tozero_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING TOZERO Image")

fig.add_subplot(rows, columns, 5)
plt.imshow(my_thresh_tozero_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING TOZERO INV Image")
```

```python
_, my_thresh_trunc_img = cv2.threshold(my_original_img, thresh=128, maxval=255, type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(14, 7))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_original_img, cmap="gray")
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_thresh_binary_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_thresh_binary_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING INV Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_thresh_tozero_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING TOZERO Image")

fig.add_subplot(rows, columns, 5)
plt.imshow(my_thresh_tozero_inv_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING TOZERO INV Image")


fig.add_subplot(rows, columns, 6)
plt.imshow(my_thresh_trunc_img, cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDING TRUNC Image")
```
