```python
width, height, channel = 300, 300, 3
img_red = np.zeros((width,height,channel),dtype=np.uint8)
img_red[:,:,0] = 255
plt.imshow(img_red)
plt.axis('off')
```

```python
width, height, channel = 300, 300, 3
img_green = np.zeros((width,height,channel),dtype=np.uint8)
img_green[:,:,1] = 255
plt.imshow(img_green)
plt.axis('off')
```

```python
width, height, channel = 300, 300, 3
img_blue = np.zeros((width,height,channel),dtype=np.uint8)
img_blue[:,:,2] = 255
plt.imshow(img_blue)
plt.axis('off')
```

```python
fig = plt.figure(figsize=(10,7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(img_red)
plt.axis('off')
plt.title("RED Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(img_green)
plt.axis('off')
plt.title("GREEN Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(img_blue)
plt.axis('off')
plt.title("BLUE Image")
```

```python
width, height, channel = 300, 420, 3
img_flag_rusia = np.zeros((width,height,channel),dtype=np.uint8)

img_flag_rusia[:100,:] = [220,220,220]

img_flag_rusia[100:200,:] = [0,0,255]

img_flag_rusia[200:300,:] = [255,0,0]

plt.imshow(img_flag_rusia)
plt.axis('off')
```

```python
mask_red = (img_flag_rusia[:, :, 0] ==255) & (img_flag_rusia[:, :, 1] ==0)

img_flag_rusia[mask_red] = [255, 255, 0]


plt.imshow(img_flag_rusia)

plt.axis('off')

```
